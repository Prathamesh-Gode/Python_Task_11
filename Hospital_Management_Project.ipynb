# Hospital Management System 

```python

class Patient:
    def __init__(self, pid, name, age, disease):
        self.pid = pid
        self.name = name
        self.age = age
        self.disease = disease

    def display_patient(self):
        print(f"Patient ID: {self.pid}")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Disease: {self.disease}")


class Doctor:
    def __init__(self, did, name, specialization):
        self.did = did
        self.name = name
        self.specialization = specialization

    def display_doctor(self):
        print(f"Doctor ID: {self.did}")
        print(f"Name: {self.name}")
        print(f"Specialization: {self.specialization}")


class Appointment:
    def __init__(self, patient, doctor, date):
        self.patient = patient
        self.doctor = doctor
        self.date = date

    def show_appointment(self):
        print("Appointment Details")
        print(f"Patient: {self.patient.name}")
        print(f"Doctor: {self.doctor.name}")
        print(f"Date: {self.date}")

class Billing:
    def __init__(self, patient, amount):
        self.patient = patient
        self.amount = amount

    def generate_bill(self):
        print("Billing Details")
        print(f"Patient Name: {self.patient.name}")
        print(f"Total Amount: ₹{self.amount}")


hospital_name = "City Care Hospital"
print(f"Welcome to {hospital_name}")

p1 = Patient(1, "Rahul", 28, "Fever")
d1 = Doctor(101, "Dr. Sharma", "Physician")

p1.display_patient()
print()
d1.display_doctor()
print()

a1 = Appointment(p1, d1, "10-Feb-2026")
a1.show_appointment()
print()

bill = Billing(p1, 1500)
bill.generate_bill()



```
